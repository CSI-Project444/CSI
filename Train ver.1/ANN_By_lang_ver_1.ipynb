```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
url='https://raw.githubusercontent.com/CSI-Project444/CSI/main/Data/Final01.csv'
df=pd.read_csv(url)
df
```

```python
from sklearn.preprocessing import StandardScaler 
from sklearn.preprocessing import OneHotEncoder
cols=['mag27','mag28', 'mag29', 'mag30', 'mag31',
       'mag33', 'mag34', 'mag35', 'mag36', 'mag37', 'mag38']
X=df[cols] 
scaler = StandardScaler()
X_scale=scaler.fit_transform(X)
X_scale = X_scale.round(10) 
y = np.array(df['Zone'])

encoder = OneHotEncoder()
y_targets = encoder.fit_transform(df[['Zone']]).toarray()
```

```python
from sklearn.model_selection import train_test_split
test_size=0.6
X_train, X_test, y_train, y_test = train_test_split(
    X, y_targets, 
    test_size=test_size, random_state=4)
```

```python
from keras.models import Sequential
from keras.layers.core import Dense
model = Sequential()
model.add(Dense(20, input_dim=11, activation='relu'))
model.add(Dense(20, input_dim=20, activation='relu'))
model.add(Dense(20, input_dim=20, activation='relu'))
model.add(Dense(4, activation='sigmoid'))
```

```python
model.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=100, verbose=1)
```

```python
results = model.evaluate(X_test, y_test)
print('Accuracy: %.3f' % results[1])
```
